```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('election_results_2024.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Margin'] = pd.to_numeric(df['Margin'], errors='coerce')
```

```python
df['Margin'] = df['Margin'].fillna(0)
```

```python
df.info()
```

```python
df['Margin'] = df['Margin'].astype('int64')
```

```python
df.describe()
```

```python
df.info()
```

```python
sns.countplot(x="Leading Party", data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Count the occurrences of each party
party_counts = df['Leading Party'].value_counts()

# Plot the pie chart
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts = ax.pie(party_counts, labels=None, autopct=None, startangle=140)

# Create legend labels with count and percentage
legend_labels = []
total_counts = party_counts.sum()
for i, party in enumerate(party_counts.index):
    count = party_counts[party]
    percentage = (count / total_counts) * 100
    legend_labels.append(f'{party}: {count} ({percentage:.1f}%)')

# Add the legend to the plot
ax.legend(wedges, legend_labels, title="Leading Party", loc="center left", bbox_to_anchor=(1, 0.5))

# Set the title and ensure the pie chart is circular
plt.title('Distribution of Leading Party')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
import matplotlib.colors as mcolors

# Count the occurrences of each party
party_counts = df['Leading Party'].value_counts()

# Generate a color map with as many colors as there are unique parties
colors = plt.cm.tab20.colors  # Use a colormap with many distinct colors

# Plot the pie chart
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts = ax.pie(party_counts, labels=None, autopct=None, startangle=140, colors=colors[:len(party_counts)])

# Create legend labels with count and percentage
legend_labels = []
total_counts = party_counts.sum()
for i, party in enumerate(party_counts.index):
    count = party_counts[party]
    percentage = (count / total_counts) * 100
    legend_labels.append(f'{party}: {count} ({percentage:.1f}%)')

# Add the legend to the plot
ax.legend(wedges, legend_labels, title="Leading Party", loc="center left", bbox_to_anchor=(1, 0.5))

# Set the title and ensure the pie chart is circular
plt.title('Distribution of Leading Party')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```
